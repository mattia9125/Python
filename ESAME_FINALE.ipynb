```python
# 2) Si chiede poi per ogni continente di trovare il numero di casi totali avvenuti in quello stesso continente;
# si chiede di non considerare eventuali locazioni che nel dataset non appartengono ad alcun continente
Filtro = Covid.dropna(subset=['continent'])
Continenti = Filtro.groupby('continent')['total_cases'].sum()
```

```python
# Sempre riguardo i casi totali, si chiede di scrivere del codice che, date due variabili contenenti i nomi di continenti, 
#ne confronti i seguenti relativi descrittori statistici: valore massimo, media, 
# e percentuale rispetto al numero dei casi totali nel mondo (calcolati anche sulle locazioni senza indicazione di continente);
continente_1 = "North America"
continente_2 = "Europe"
filtro = Covid[(Covid['continent'] == continente_1) | (Covid['continent'] == continente_2)]
```

```python
# Filtro per singoli continenti
continente_1_filtro = filtro[filtro['continent'] == continente_1]
continente_2_filtro = filtro[filtro['continent'] == continente_2]
```

```python
# Max continente 1
max_continente_1 = continente_1_filtro['total_cases'].max()
max_continente_1
```

```python
# Max continente 2
max_continente_2 = continente_2_filtro['total_cases'].max()
max_continente_2
```

```python
# Media continente 1
media_continente_1 = continente_1_filtro['total_cases'].mean().round(2)
media_continente_1
```

```python
# Media continente 2
media_continente_2 = continente_2_filtro['total_cases'].mean().round(2)
media_continente_2
```

```python
# Percentuale dei casi totali dei due continente 1  rispetto ai casi totali nel mondo
toale_casi_mondo = Covid['total_cases'].sum()
percentuale_cont_1 = (continente_1_filtro['total_cases'].sum() / toale_casi_mondo) * 100
print(f"{percentuale_cont_1:.2f}%")
```

```python
# Percentuale dei casi totali dei due continente 2 rispetto ai casi totali nel mondo
percentuale_cont_2 = (continente_2_filtro['total_cases'].sum() / toale_casi_mondo) * 100
print(f"{percentuale_cont_2:.2f}%")
```

```python
#Selezionare i dati relativi all'Italia nel 2022.  
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

Covid['date'] = pd.to_datetime(Covid['date'])
italy_2022 = Covid.loc[(Covid['location'] == 'Italy') & (Covid['date'].dt.year== 2022)]
italy_2022
```

```python
# Mostrare con un grafico adeguato l'evoluzione del casi totali rispetto alla data.
sns.lineplot(x='date', y='total_cases', data=italy_2022)
plt.xlabel('Data')
plt.ylabel('Casi Totali')
plt.title('Casi Totali Italia 2022')
plt.grid(True)
plt.xticks(rotation=45)  
plt.tight_layout()
plt.show()
```

```python
# Mostrare poi con un grafico adeguato il numero di nuovi casi rispetto alla data (filtrare i dati se necessario).
sns.lineplot(x='date', y='new_cases', data=italy_2022)
plt.xlabel('Data')
plt.ylabel('Nuovi Casi')
plt.title('Andamento nuovi caasi Italia 2022')
plt.grid(True)
plt.xticks(rotation=45)  
plt.tight_layout()
plt.show()
```

```python
# Mostrare infine un grafico che mostra l'andamento della somma cumulativa nuovi casi del 2022
italy_2022.loc[:,'nuovi_casi_cum'] = italy_2022['new_cases'].cumsum()
```

```python
sns.lineplot(x='date', y='nuovi_casi_cum', data=italy_2022)
plt.xlabel('Data')
plt.ylabel('Nuovi Casi Cum')
plt.title('Andamento Cumulativo nuovi casi Italia 2022')
plt.grid(True)
plt.xticks(rotation=45)  
plt.tight_layout()
plt.show()
```

L'andamento del curva dei casi totali e dei nuovi casi è molto simile. Entranbe sono delle curve in crescita. La differenza la si evidenzia nella stagionalità. Infatti, i nuovi casi aumentano in modo esponenziale durante il periodo invernale, mentre si può notare una certa stabilità tra maggio e luglio.

```python
Confronto_nazioni = Covid.loc[(Covid['location'] == 'Italy') | (Covid['location'] == 'Germany') | (Covid['location'] == 'France')]
```

```python
Confronto_nazioni
```

```python
Conf_nazio_2022_2023 = Confronto_nazioni.loc[(Confronto_nazioni['date'].dt.year == 2022 ) | (Confronto_nazioni['date'].dt.year == 2023)]
```

```python
data_inizio = pd.to_datetime('2022-05-01')
data_fine = pd.to_datetime('2023-04-30')
Conf_nazio_mag_apr = Conf_nazio_2022_2023.loc[(Conf_nazio_2022_2023 ['date'] >= data_inizio) & (Conf_nazio_2022_2023 ['date'] <= data_fine)]
```

```python
Conf_nazio_mag_apr 
```

```python
sns.boxplot(x = "location", y = "icu_patients",  data= Conf_nazio_mag_apr)
plt.xticks(rotation=45)
plt.title("Pazienti in terapia intensiva in Italia, Germania e Francia (Maggio 2022 - Aprile 2023)")
plt.show()
```

Nella distibuzione generale dei dati si evince che non compaiono Outliers (Valori anomali) che potrebbero inficiare l'analisi.La grandezza delle scatole di Francia e Germania sottolineano una maggiore variabilità dei valori rispetto alla mediana, cosa che non si riscontra per quanto riguarda l'italia dove si evidenzia poca variabilità dei dati rispetto alla mediana. Si evidenzia, inoltre, che per i pazienti in ICU in Italia si ha valore mediamo molto più basso rispetto a Francia e Germania. Mentre per Francia e Germania questo è quasi parititetico, anche se è tendezialmente più alto quello tedesco. Per quanto riguarda Francia e Germania ci troviamo di fronte ad una asimmetria, in quanto il baffo superioriore è nettamente più lungo del baffo infeririore e questo evidenza una maggiore disperzione dei dati, ovvero i valori più grandi presentano una maggiore dispersione.
